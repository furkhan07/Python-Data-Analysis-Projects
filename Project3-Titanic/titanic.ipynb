# DATA CLEANING 

1. Remove duplicates
2. Mean , Median , Mode
3. astype()/to_datetime()

```python
df.duplicated().any()
#  NO DUPLICATES
```

MISSING DATA COLUMNS   
> age - float - mean  
> fare - float - mean  
> cabin(too many null values so we drop this column)  
> embarked - object - mode 

```python
#Finding the replacement values for null values
a_m=df['age'].mean() #OR directly df['age']=df['age'].fillna(df['age'].mean)
f_m=df['fare'].mean()
e_m=df['embarked'].mode()[0] #return type of mode is Series
```

```python
df['age']=df['age'].fillna(a_m)
df['fare']=df['fare'].fillna(f_m)
df['embarked']=df['embarked'].fillna(e_m)
```

```python
df.info()
```

```python
#Drop Cabin Column as more than 50% of data is missing
df=df.drop(columns=['cabin'])
df.info()
```
