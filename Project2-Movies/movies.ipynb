```python
df['Age']
```

```python
df['age']=df['Age'].str.split('+',expand=True)[0]
df
df=df.drop(columns=['Age'])
# df
```

```python
#since age is of datatype object we are taking MODE
a_m=df['age'].mode()[0]
df['age']=df['age'].fillna(a_m)
# df.info()
```

```python
#IMDb

i_m=df['IMDb'].mean().round(1)
df['IMDb']=df['IMDb'].fillna(i_m)
# df.info()
```

TRANSFORMATION  
since IMDb and Rotten Tomatoes both are ratings , to convert we have to convert them to single type or same scale  
 RottenTomatoes --> strip('%') then divide by 10 --> we'll get the range 0-10

```python
#Rotten Tomatoes    #using mode
df['Rotten Tomatoes']  
#the values are present with % so we have to remove it , since we have to compare IMDb and Rotten Tomatoes we have to keep both in same range
df['Rotten Tomatoes']=(df['Rotten Tomatoes'].str.strip('%').astype(float)/10).round(1)
r_m=df['Rotten Tomatoes'].mode()[0]
df['Rotten Tomatoes']=df['Rotten Tomatoes'].fillna(r_m)
df
```

```python
#Directors , Genres , Country , Language --> all of these have missing values and we cannot take mean, median or mode so we have to assign UNKNOWN

for i in ['Directors','Genres','Country','Language']:
    df[i]=df[i].fillna('unknown')
```

```python
#Runtime

rt_m=df['Runtime'].mean()
df['Runtime']=df['Runtime'].fillna(rt_m).round(1)
df.info()
```

```python
# Single cell has multiple values so we have to separate them
```

```python
df['runtime category']=pd.cut(df['Runtime'],bins=[0,120,180,240],labels=['short','mid','long'])
df.head()
```

**melt()**

Converts Rows to Columns : Converting the array to series which is further converted to dataframe
> Rows will Increase , Columns will Decrease

** After melting we can apply groupby for netflix,hulu,prime videos,disney as it applies on rows 

```python
plts_df=pd.melt(df,id_vars=['Title','IMDb','Rotten Tomatoes'],value_vars=['Netflix','Hulu','Prime Video','Disney+'],var_name='Platforms',value_name='Availability')
plts_df=plts_df.loc[plts_df['Availability']==1]   
plts_df
#1 means only the ones which are available in that platform , removing the 0's which are not available in the platforms
```

```python
#Since we have taken for Availability=1 , the entire df will have 1 so remove the column 
plts_df=plts_df.drop(columns=['Availability'])

plts_df
```

Diff b/w Pivot and Pivot Table methods in Pandas

**PIVOT & PIVOT TABLE**    
> PIVOT will throw Error if we have duplicates in the dataframe (here duplicates refer to the Index&Column combination is repeated )  
> PIVOT-TABLE will not throw error for Duplicates , it will group them and apply aggregate function on them   

**PIVOT & MELT**    
> Converts Rows to Columns (Long format to Wide format)   -> BECOMES WIDER THAN THE ORIGINAL TABLE , Length is not affected much  
> Converts Columns to Rows (Wide Format to Long Format)   -> BECOMES LONGER THAN THE ORIGINAL TABLE , Width is not affected much
